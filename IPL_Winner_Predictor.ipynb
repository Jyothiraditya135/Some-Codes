```python
import pandas as pd
import numpy as np
```

```python
raw_df = pd.read_csv('/content/IPL_train_ dataset 2008-2021.csv') #the path of the csv file may be different for you change it accordingly
```

```python
raw_df.replace(to_replace="RCB", value="Royal Challengers Bangalore",inplace=True)
raw_df.replace(to_replace="Capitals", value="Delhi Capitals",inplace=True)
raw_df.replace(to_replace="KKR", value="Kolkata Knight Riders",inplace=True)
raw_df.replace(to_replace="Punjab", value="Punjab Kings",inplace=True)
raw_df.replace(to_replace="Mumbai", value="Mumbai Indians",inplace=True)
raw_df.replace(to_replace="Royals", value="Rajasthan Royals",inplace=True)
raw_df.replace(to_replace="Super", value="Chennai Super Kings",inplace=True)
raw_df.replace(to_replace="Sunrisers", value="Sunrisers Hyderabad",inplace=True)
raw_df.replace(to_replace="Kings", value="Kings XI Punjab",inplace=True)
raw_df.replace(to_replace="Daredevils", value="Delhi Daredevils",inplace=True)
raw_df.replace(to_replace=["Supergiant","Supergiants"], value="Rising Pune Supergiants",inplace=True)
raw_df.replace(to_replace="Guj", value="Gujarat Lions",inplace=True)
raw_df.replace(to_replace="Warriors", value="Pune Warriors",inplace=True)
raw_df.replace(to_replace="Chargers", value="Deccan Chargers",inplace=True)
raw_df.replace(to_replace="Kochi", value="Kochi Tuskers Kerala",inplace=True)
```

```python
raw_df.drop(raw_df.index[(raw_df["winner"] == "Tied")],axis=0,inplace=True)     #Tied matches
```

```python
raw_df.drop(raw_df.index[(raw_df["winner"] == "Match")],axis=0,inplace=True)    #Match abandoned without a ball
```

```python
raw_df.drop(raw_df.index[(raw_df["winner"] == "No")],axis=0,inplace=True)       #No result
```

```python
raw_df.dropna(subset=['team1_score', 'team2_score'], inplace=True)
```

```python
raw_df.columns = raw_df.columns.str.strip()
```

```python
team_names = ['Mumbai Indians', 'Chennai Super Kings', 'Kolkata Knight Riders', 'Punjab Kings', 'Royal Challengers Bangalore', 'Delhi Capitals', 'Rajasthan Royals', 'Sunrisers Hyderabad', 'Kings XI Punjab', 'Delhi Daredevils', 'Gujarat Lions', 'Rising Pune Supergiant', 'Rising Pune Supergiants', 'Pune Warriors', 'Deccan Chargers', 'Kochi Tuskers Kerala']

team_data = {}

for team in team_names:
  team_df = raw_df.loc[((raw_df['team1'] == team) | (raw_df['team2'] == team))]

  team_df.drop(columns=['margin', 'man_of_the_match', 'place', 'year'], axis=1, inplace=True)

  team_df.reset_index(inplace=True)

  team_score = []
  for i in range(len(team_df)):
    if team_df['team1'].values[i] == team:
        team_score.append(team_df['team1_score'].values[i])
    else:
        team_score.append(team_df['team2_score'].values[i])

  team_df.drop(columns=['team1_score', 'team2_score', 'index'], axis=1, inplace=True)

  for i in range(len(team_df)):
    if team_df['toss_winner'].values[i] == team:
        team_df['toss_winner'].values[i] = 1
    else:
        team_df['toss_winner'].values[i] = 0

  for i in range(len(team_df)):
    if team_df['winner'].values[i] == team:
      team_df['winner'].values[i] = 1
    else:
      team_df['winner'].values[i] = 0

  opponent_team = []
  for i in range(len(team_df)):
    if team_df['team1'].values[i] != team:
        opponent_team.append(team_df['team1'].values[i])
    else:
        opponent_team.append(team_df['team2'].values[i])

  team_df['Opponent_Team'] = opponent_team

  team_df.drop(columns=['team1', 'team2'], axis=1, inplace=True)

  team_data[team] = team_df
```

```python
RCB_encoded = pd.get_dummies(team_data['Royal Challengers Bangalore'], columns=team_data['Royal Challengers Bangalore'].columns.difference(['winner', 'toss_winner', 'Team_Score', 'year']))
MI_encoded = pd.get_dummies(team_data['Mumbai Indians'], columns=team_data['Mumbai Indians'].columns.difference(['winner', 'toss_winner', 'Team_Score', 'year']))
CSK_encoded = pd.get_dummies(team_data['Chennai Super Kings'], columns=team_data['Chennai Super Kings'].columns.difference(['winner', 'toss_winner', 'Team_Score', 'year']))
KKR_encoded = pd.get_dummies(team_data['Kolkata Knight Riders'], columns=team_data['Kolkata Knight Riders'].columns.difference(['winner', 'toss_winner', 'Team_Score', 'year']))
PK_encoded = pd.get_dummies(team_data['Punjab Kings'], columns=team_data['Punjab Kings'].columns.difference(['winner', 'toss_winner', 'Team_Score', 'year']))
DC_encoded = pd.get_dummies(team_data['Delhi Capitals'], columns=team_data['Delhi Capitals'].columns.difference(['winner', 'toss_winner', 'Team_Score', 'year']))
RR_encoded = pd.get_dummies(team_data['Rajasthan Royals'], columns=team_data['Rajasthan Royals'].columns.difference(['winner', 'toss_winner', 'Team_Score', 'year']))
SRH_encoded = pd.get_dummies(team_data['Sunrisers Hyderabad'], columns=team_data['Sunrisers Hyderabad'].columns.difference(['winner', 'toss_winner', 'Team_Score', 'year']))
K1P_encoded = pd.get_dummies(team_data['Kings XI Punjab'], columns=team_data['Kings XI Punjab'].columns.difference(['winner', 'toss_winner', 'Team_Score', 'year']))
DD_encoded = pd.get_dummies(team_data['Delhi Daredevils'], columns=team_data['Delhi Daredevils'].columns.difference(['winner', 'toss_winner', 'Team_Score', 'year']))
GL_encoded = pd.get_dummies(team_data['Gujarat Lions'], columns=team_data['Gujarat Lions'].columns.difference(['winner', 'toss_winner', 'Team_Score', 'year']))
RPS_encoded = pd.get_dummies(team_data['Rising Pune Supergiant'].append(team_data['Rising Pune Supergiant']), columns=team_data['Royal Challengers Bangalore'].columns.difference(['winner', 'toss_winner', 'Team_Score', 'year']))
PW_encoded = pd.get_dummies(team_data['Pune Warriors'], columns=team_data['Pune Warriors'].columns.difference(['winner', 'toss_winner', 'Team_Score', 'year']))
DCH_encoded = pd.get_dummies(team_data['Deccan Chargers'], columns=team_data['Deccan Chargers'].columns.difference(['winner', 'toss_winner', 'Team_Score', 'year']))
KTK_encoded = pd.get_dummies(team_data['Kochi Tuskers Kerala'], columns=team_data['Kochi Tuskers Kerala'].columns.difference(['winner', 'toss_winner', 'Team_Score', 'year']))
```

```python
team_dfs = [RCB_encoded, MI_encoded, CSK_encoded, KKR_encoded, PK_encoded, DC_encoded, RR_encoded, SRH_encoded, K1P_encoded, DD_encoded, GL_encoded, RPS_encoded, PW_encoded, DCH_encoded, KTK_encoded]

train_ratio = 0.8
X_train_list = []
Y_train_score_list = []
Y_train_list = []
X_test_list = []
Y_test_score_list = []
Y_test_list = []

for team_df in team_dfs:
    total_rows = len(team_df)
    train_rows = int(total_rows * train_ratio)

    Y_train = team_df[:train_rows][['winner']].copy()
    Y_train_score = team_df[:train_rows][['Team_Score']].copy()
    X_train = team_df[:train_rows].drop(['winner', 'Team_Score'], axis=1)
    X_train_list.append(X_train)
    Y_train_score_list.append(Y_train_score)
    Y_train_list.append(Y_train)

    Y_test = team_df[train_rows:][['winner']].copy()
    Y_test_score = team_df[:train_rows][['Team_Score']].copy()
    X_test = team_df[:train_rows].drop(['winner'], axis=1)
    X_test = team_df[train_rows:].drop(['winner', 'Team_Score'], axis=1)
    X_test_list.append(X_test)
    Y_test_score_list.append(Y_test_score)
    Y_test_list.append(Y_test)
```

```python
X_train_list_np = [df.values for df in X_train_list]
Y_train_score_list_np = [df.values for df in Y_train_score_list]
Y_train_list_np = [df.values for df in Y_train_list]
X_test_list_np = [df.values for df in X_test_list]
Y_test_score_list_np = [df.values for df in Y_test_score_list]
Y_test_list_np = [df.values for df in Y_test_list]
```

```python
params_list = [np.random.randn(*(len(df.columns)-1,1)) for df in team_dfs]
params_list[1] = np.random.randn(*(49,1))
```

```python
def lin_reg_model(xtr, ytr, pars, alpha, epochs):
  for _ in range(epochs):
    y_pred = np.matmul(xtr, pars)
    error = (y_pred - ytr)

    cost_fnc = 0.5 * np.mean(error**2)
    grads = np.matmul(xtr.T, error)/np.size(ytr)
    pars = pars - alpha*(grads)

  return pars
```
